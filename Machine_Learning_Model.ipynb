    #2. Mã hóa dữ liệu

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
X = df3.iloc[:,:-1].values
X
```

```python
Y = df3.iloc[:,-1:].values
Y
```

```python
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse=False),[4,5])],remainder = 'passthrough')
X = ct.fit_transform(X)
```

```python
X
```

    #3 Xử lý dữ liệu Nan


```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean') # Chuyển dữ liệu Null thành trung bình của từng cột
imputer.fit(X)
X = imputer.transform(X)
X
```

# Machine Learning Model
    1. Chia mô hình ( 80% cho việc train , 20% cho việc test )
    2. Xử lý dữ liệu ngoại lai 

                                            #1 Chia mô hình

```python
len(X),len(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y ,train_size = 0.8,test_size = 0.2,random_state = 0)
```

```python
len(X_train),len(Y_train) 
```

```python
len(X_test),len(Y_test)
```

```python
X_train, X_test,Y_train,Y_test
```

                                            #2 Xử lý dữ liệu ngoại lai

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train,X_test
```

                                    #3 Xây dựng mô hình máy học

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state = 1)
dt_model.fit(X_train, Y_train)
price_predict = dt_model.predict(X_test)
len(price_predict),len(Y_test)
```

```python
df_ML = pd.DataFrame(list(zip(Y_test,price_predict)),columns = ['Price','Price Predict'])
```

```python
df_ML 
```

```python
df_ML['new_col'] = df_ML['Price Predict'] - df_ML['Price']
df_ML
```

```python
def bao_cao(do_chenh_lech):
    dem = 0
    for x in df_ML['new_col'].values:
        if abs(x) <= do_chenh_lech:
            dem+=1
            percent = round(dem/len(df_ML)*100,2)
    if do_chenh_lech == 0:
        print('Mô hình dư đoán chính xác giá nhà ở mức {} %'.format(percent))
    else:
        print('Mô hình dư đoán giá nhà ở dưới mức {} là : {} %'.format(do_chenh_lech,percent))
```

```python
do_chenh_lech = [0,1000000,100000000,1000000000]
for x in do_chenh_lech:
    bao_cao(x)
```
